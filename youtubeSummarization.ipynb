```python
!pip install transformers pytube youtube-transcript-api

from transformers import pipeline
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi

# Load the YouTube video transcript
video_id = "dQw4w9WgXcQ"  # Replace with the actual video ID
transcript = YouTubeTranscriptApi.get_transcript(video_id)
transcript_text = " ".join([entry["text"] for entry in transcript])

# Initialize the summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Perform summarization
summary = summarizer(transcript_text, max_length=130, min_length=30, do_sample=False)

# Print the summary
print(summary[0]["summary_text"])
```
